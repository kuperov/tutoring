```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates

# Integer hours of the day (0 to 23)
# Instead of simple integers, we'll create datetime objects to represent the hours
# An arbitrary date (July 20, 2025) is used, as only the time component is relevant for the plot.
start_time = datetime.datetime(2025, 7, 20, 0, 0, 0)
time_objects = [start_time + datetime.timedelta(days=h) for h in range(7)]

# maximum daily temperatures
temp = np.array([15., 17, 18, 20, 19, 17, 17,]) # 18, 17, 17.])

# Create the plot
plt.figure(figsize=(4, 3)) # Set figure size to (4, 3) as requested

# Plot the data in black with circles as markers, using datetime objects for x-axis
plt.plot(time_objects, temp, color='black', marker='o', linestyle='-', linewidth=1, markersize=5)

# Add title and labels
plt.title('Forecast Maximum Daily Temperatures', color='black')
plt.ylabel('Temperature (°C)', color='black')

# Format the x-axis as time
# Use HourLocator to set ticks at specific intervals (e.g., every 4 hours)
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))
# Format as day name
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%A'))

# Rotate date labels for better readability if they overlap
plt.gcf().autofmt_xdate()

# Add a grid for better readability, in light grey
plt.grid(True, linestyle='--', alpha=0.7, color='grey')

plt.ylim([14, 21])

plt.savefig('46_temp.pdf', bbox_inches='tight')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/IDCJAC0002_066037_Data1.csv')
df.columns = ['code', 'station', 'year', 'month', 'max_temp', 'quality']
df['day'] = 1
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

#df = df.groupby('date').agg({'max_temp': 'mean'}).reset_index()

fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(df['date'], df['max_temp'], linestyle='-', color='black', linewidth=1)
ax.set_ylabel('Temperature (°C)', fontsize=12)
ax.set_xlabel('Date', fontsize=12)
ax.set_title('Maximum Temperature in Sydney Airport', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7, color='grey')

# plt.savefig('fig/line_max_temp.pdf', bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
df_jun = df[df['month'] == 6]
df_jan = df[df['month'] == 1]
ax.plot(df_jan['date'], df_jan['max_temp'], linestyle=':', color='red', linewidth=1, label='January')
ax.plot(df_jun['date'], df_jun['max_temp'], linestyle='-', color='black', linewidth=1, label='June')
ax.legend(fontsize=12, frameon=True)
ax.set_ylabel('Temperature (°C)', fontsize=12)
ax.set_title('Average Maximum Monthly Temperature, Sydney Airport', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.5, color='grey')
ax.set_yticks(range(14, 32, 2))
fig.tight_layout()

fig.savefig('fig/line_max_monthly_temp_syd.pdf', bbox_inches='tight')
```

```python
rs = pd.read_csv('data/abs_monthly_retail_sales.csv')
rs['month'] = pd.to_datetime(rs['month'])
rs = rs.set_index('month')
rs.tail()
```

```python
# plot data in rs
fig, ax = plt.subplots(figsize=(12, 4))
rs2019 = rs[rs.index >= pd.to_datetime('2019-01-01')]
ax.plot(rs2019.index, 1e-3*rs2019['total'], color='black', linewidth=1)
ax.set_ylabel('Total Retail Sales ($ billions)', fontsize=12)
ax.set_title('Monthly Retail Sales, Australia', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.5, color='grey')
fig.tight_layout()

fig.savefig('fig/line_retail_sales.pdf', bbox_inches='tight')
```
